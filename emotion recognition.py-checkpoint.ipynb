#real time recognition

```python
import cv2
from deepface import DeepFace
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
        raise IOError("cannot open webcame")
        
while True:
    ret,frame = cap.read()
    result = DeepFace.analyze(frame, actions = ['emotion'])
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray,1.1,4)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame, (x,y),(x+w, y+h), (0, 255, 0),2)
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(frame,
               result[0]['dominant_emotion'],
               (50, 50),
               font, 3,
               (0, 0, 225),
               2,
               cv2.LINE_4)
    cv2.imshow('Demo Video', frame)
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import speech_recognition as sr
from textblob import TextBlob

# Initialize the recognizer
r = sr.Recognizer()

# Function to perform emotion detection on recognized speech
def detect_emotion(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity

    if sentiment > 0:
        return "Positive"
    elif sentiment < 0:
        return "Negative"
    else:
        return "Neutral"

# Function to listen and recognize speech in real-time
def listen():
    with sr.Microphone() as source:
        print("Listening...")

        # Adjust the ambient noise threshold if necessary
        r.adjust_for_ambient_noise(source)

        # Continuously listen to the microphone
        while True:
            audio = r.listen(source)

            try:
                # Perform speech recognition
                text = r.recognize_google(audio)
                print("You said:", text)

                # Perform emotion detection
                emotion = detect_emotion(text)
                print("Emotion:", emotion)
            except sr.UnknownValueError:
                print("Unable to recognize speech")
            except sr.RequestError as e:
                print("Could not request results from Google Speech Recognition service; {0}".format(e))

# Call the function to start real-time speech recognition
listen()
```
